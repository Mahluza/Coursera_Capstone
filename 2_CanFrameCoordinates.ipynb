CONJOINING WIKIPEDIA DATA FRAME WITH COORDINATES FOR EACH POSTAL CODE 

```python
import pandas as pd
```

```python
#Import dataframe created from wiki page
cpc_df3 = pd.read_csv('/resources/data/Canada.csv')
```

```python
cpc_df3.head()
```

```python
#Drop 'Unnamed: 0' column
cpc_df3.drop('Unnamed: 0', 1, inplace=True)
```

```python
#Create dataframe from coordinate data 
latlon_df = pd.read_csv('/resources/data/Geospatial_Coordinates.csv')
latlon_df.rename(columns={'Postal Code': 'Postcode'}, inplace=True)
```

```python
latlon_df.head()
```

```python
#Merge the two frames 
cpc_df4 = pd.merge(cpc_df3, latlon_df, on='Postcode')
```
